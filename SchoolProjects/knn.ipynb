# kNN Prediction

## Case A

Using included autos dataset (autos-clean.csv) predict the city-mpg of gas powered cars. Following features should be used in prediction:
- engine-size
- horsepower
- curb-weight
- length
- highway-mpg

Prediction algorithm in this case should be kNN with k value of 5. 
Use train_test_split function with test_size=0.3 and random_state=100.

Evaluate the prediction model by calculating values of R-squared, MAE, MSE and RMSE. 
Draw a scatter plot from test set prediction and true values. 

```python
# Imports
from sklearn.neighbors import KNeighborsClassifier # For k-NN
from sklearn.model_selection import train_test_split
from sklearn import metrics # For validation
import pandas as pd # For data
import numpy as np
import matplotlib.pyplot as plt # for scatter plot
```

```python
# Data
data = pd.read_csv("autos_clean.csv")

# Data split
y = data["city-mpg"]
X = data[["engine-size", "horsepower", "curb-weight", "length", "highway-mpg"]]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 100)
```

```python
# Case A
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)

# Metrics
mae = metrics.mean_absolute_error(y_test, y_pred)
mse = metrics.mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)

# Prints
print('Mean Absolute Error:', mae)
print('Mean Square Error:', mse)
print('Root Mean Square Error:', rmse)

# Plot
X_values = np.arange(X_test.shape[0])

plt.figure(figsize=(10,8))
plt.scatter(X_values, y_test.values, marker='*')
plt.scatter(X_values, y_pred, marker='+')
plt.title("Test values (*) and predicted values (+)")
plt.grid()
plt.show()
```


## Definition Case B

Using included autos dataset (autos-clean.csv) predict feature drive-wheels
using features: 'engine-size', 'curb-weight'

Evaluate the predicted model using accuracy of the test_set.
Visualize the results using confusion matrix.

```python
# Autos and import already done...

y2 = data["drive-wheels"]
X2 = data[["engine-size", "curb-weight"]]

X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y2, test_size = 0.3, random_state = 100)

knn2 = KNeighborsClassifier(n_neighbors=5)
knn2.fit(X_train2, y_train2)
y_pred2 = knn2.predict(X_test2)

# Metrics
confusionMatrix = metrics.confusion_matrix(y_true=y_test2, y_pred=y_pred2)

# Plot
confusionMatrixDisplay = metrics.ConfusionMatrixDisplay(confusion_matrix = confusionMatrix)
confusionMatrixDisplay.plot()
plt.show()
```
